```python
# Step 2: Load the model and Gradio
from transformers import pipeline
import gradio as gr
import tempfile
# Load a summarization model
summarizer = pipeline("summarization", model="Falconsai/text_summarization")
# Define summarization function
def summarize_text(text):
    summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
    return summary[0]['summary_text']
def summarize_and_export(text):
    if not text.strip():
        return "Please enter some text to summarize.", None
    summary = summarizer(text, max_length=130, min_length=30, do_sample=False)[0]['summary_text']
    # Create a temporary file to store the summary
    file_path = "/content/summary.txt"
    with open(file_path, "w") as f:
        f.write(summary)
    # Return summary and file path for download
    return summary, file_path
# Step 3: Create Gradio interface

interface = gr.Interface(
    fn=summarize_and_export,
    inputs=gr.Textbox(lines=10, placeholder="Enter text to summarize..."),
    outputs=[
        gr.Textbox(lines=10, label="Summary"),
        gr.File(label="Download Summary")
    ],
    title="Text Summarization App with Export",
    description="Enter text, generate a summary with a Hugging Face model, and download the summary as a file."
)
# Step 4: Launch the interface

interface.launch()
```
